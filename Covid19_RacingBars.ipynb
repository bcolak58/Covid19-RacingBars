**Covid19 Chart of Countries using RacingBars**

```python
import pandas as pd
```

```python
df=pd.read_csv('corona_dat.csv')
```

```python
df.head()
```

```python
#birden fazla parametre olacak ise 2 tane köşeli paramteze alınır
df2=df[['date','China','Italy','Brazil','Spain','US','United Kingdom','Turkey']]
```

```python
df2.sample()
```

```python
df2.set_index('date',inplace=True) #sol başta yer alan sayısal alan yerine index tarih yapıldı
```

```python
df2.sample()
```

```python
toplam=df2.cumsum(axis=0)
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(toplam,filename='covid19.mp4',figsize=(8,6),title='Ülke Covid19 verileri')
```
